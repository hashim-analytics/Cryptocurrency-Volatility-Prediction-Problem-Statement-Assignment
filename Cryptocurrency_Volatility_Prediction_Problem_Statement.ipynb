```python
import zipfile
import pandas as pd

# Zip file ka path
zip_path = "dataset.csv.zip"

# Extract the CSV
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall()  # default current folder me extract karega
    print("Files in zip:", zip_ref.namelist())  # ye batayega CSV ka exact naam

# Ab CSV load karo (jo zip me file ka exact naam hai use use karo)
df = pd.read_csv(zip_ref.namelist()[0])
df.head()
```

```python
# Step 1: Basic Data Inspection

# 1️⃣ Dataset shape (rows, columns)
print("Dataset Shape:", df.shape)

# 2️⃣ First 5 rows
print("First 5 rows:")
display(df.head())

# 3️⃣ Column names and data types
print("\nColumn Data Types:")
print(df.dtypes)

# 4️⃣ Missing values per column
print("\nMissing Values per Column:")
print(df.isnull().sum())

# 5️⃣ Basic statistics for numeric columns
print("\nBasic Statistics:")
display(df.describe())

# 6️⃣ Check unique cryptocurrencies
print("\nUnique Cryptocurrencies:", df['crypto_name'].unique())
```

```python
#Step 2: Data Preprocessing

# 1️⃣ Drop redundant column
df = df.drop(columns=['Unnamed: 0', 'timestamp'])

# 2️⃣ Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'])

# 3️⃣ Sort by crypto_name and date
df = df.sort_values(['crypto_name', 'date']).reset_index(drop=True)

# 4️⃣ Check updated dataframe
print("Updated Dataset Shape:", df.shape)
display(df.head())
```

```python
# Step 3: Feature Engineering

# 1️⃣ Daily Returns
df['return'] = df.groupby('crypto_name')['close'].pct_change()

# 2️⃣ Rolling Volatility (7-day)
df['volatility_7d'] = df.groupby('crypto_name')['return'].rolling(window=7).std().reset_index(0, drop=True)

# 3️⃣ Liquidity Ratio
df['liquidity'] = df['volume'] / df['marketCap']

# 4️⃣ 20-day Moving Average & Bollinger Bands
df['MA20'] = df.groupby('crypto_name')['close'].transform(lambda x: x.rolling(20).mean())
df['STD20'] = df.groupby('crypto_name')['close'].transform(lambda x: x.rolling(20).std())
df['Upper_BB'] = df['MA20'] + 2 * df['STD20']
df['Lower_BB'] = df['MA20'] - 2 * df['STD20']

# 5️⃣ Drop rows with NaN from rolling features
df = df.dropna().reset_index(drop=True)

# 6️⃣ Check new features
print("Dataset after Feature Engineering:")
display(df.head())
print("Updated Shape:", df.shape)
```

```python
# Step 4: Exploratory Data Analysis (EDA)

import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[['return','volatility_7d','liquidity','MA20','Upper_BB','Lower_BB']].corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()

# 2️⃣ Volatility Trend for a Cryptocurrency
crypto = 'Bitcoin'  # Change any crypto from df['crypto_name'].unique()
crypto_df = df[df['crypto_name'] == crypto]

plt.figure(figsize=(12,5))
plt.plot(crypto_df['date'], crypto_df['volatility_7d'], label='7-day Volatility')
plt.title(f"{crypto} Volatility Over Time")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.show()

# 3️⃣ Distribution of Daily Returns
plt.figure(figsize=(10,5))
sns.histplot(crypto_df['return'], bins=50, kde=True)
plt.title(f"{crypto} Daily Returns Distribution")
plt.xlabel("Daily Return")
plt.show()
```

2️⃣ LLD Enhancements

Use functions for each module

Makes design more readable and professional

```python
# Compute daily returns
def compute_returns(df):
    df['return'] = df.groupby('crypto_name')['close'].pct_change()
    return df

# Compute rolling volatility
def compute_rolling_volatility(df, window=7):
    df['volatility_7d'] = df.groupby('crypto_name')['return'].rolling(window).std().reset_index(0, drop=True)
    return df

# Compute liquidity ratio
def compute_liquidity(df):
    df['liquidity'] = df['volume'] / df['marketCap']
    return df

# Apply all
df = compute_returns(df)
df = compute_rolling_volatility(df)
df = compute_liquidity(df)
```
